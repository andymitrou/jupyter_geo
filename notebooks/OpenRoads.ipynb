```python
import os
import geopandas
import folium
```

```python
#Define input 
#***IMPORTANT****
# Need to change the following paths
input_shapefile = "../data/TQ_RoadLink.shp"
output_geojson = "../data/TQ_RoadLink.json"
# Open the file and read the first 100 records
data = geopandas.read_file(input_shapefile)
subset = data.head(100)

# Reproject the subset to WGS 84 in order to be able to be visualised through Follium.
subset_wgs84 = subset.to_crs(epsg=4326)

#Save it in GeoJSON
subset_wgs84.to_file(output_geojson,driver="GeoJSON")
print("File saved")

m = folium.Map(zoom_start=4,location=[54.5263833,-7.2406521])
m

mapLayer = folium.GeoJson(output_geojson, name='geojson').add_to(m)
m.fit_bounds(mapLayer.get_bounds())
m

```
